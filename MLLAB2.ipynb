## Identify and handle missing values

### Subtask:
Check for any missing values in the dataset and apply an appropriate strategy to deal with them, such as imputation or removal.


```python
print("Missing values per column:")
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df['Purchased_Numeric'] = df['Purchased'].map({'No': 0, 'Yes': 1})
df['Annual_Salary_K'] = df['Salary'] / 1000
print("Data transformations complete. Displaying the first few rows of the DataFrame to verify:")
display(df.head())

print("\nDisplaying a concise summary of the DataFrame to check updated data types and non-null counts:")
df.info()
```

**Reasoning**:
The previous code generated a `FutureWarning` due to the use of `inplace=True` with chained assignment. To address this and adhere to best practices for `pandas` operations, I will modify the `.fillna()` calls to explicitly assign the results back to the DataFrame columns, thereby removing the `inplace=True` argument. The rest of the transformations will remain the same.



```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Purchased_Numeric'] = df['Purchased'].map({'No': 0, 'Yes': 1})
df['Annual_Salary_K'] = df['Salary'] / 1000
print("Data transformations complete. Displaying the first few rows of the DataFrame to verify:")
display(df.head())

print("\nDisplaying a concise summary of the DataFrame to check updated data types and non-null counts:")
df.info()
```

```python
print("1. Selected columns 'Country', 'Age', and 'Annual_Salary_K':")
display(df[['Country', 'Age', 'Annual_Salary_K']])

print("\n2. Rows where 'Age' is greater than 30:")
display(df[df['Age'] > 30])

print("\n3. Rows where 'Country' is 'France' AND 'Purchased_Numeric' is 1 (Yes):")
display(df[(df['Country'] == 'France') & (df['Purchased_Numeric'] == 1)])

print("\n4. Rows where 'Salary' is greater than 60000 AND less than 70000:")
display(df[(df['Salary'] > 60000) & (df['Salary'] < 70000)])
```
